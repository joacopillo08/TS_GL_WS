Se programo una funci칩n que genera se침ales senoidales y que permite parametrizar la frecuencia de muestreo, la cantidad de muestras por ciclo, la amplitud y la frecuencia. 

```python
# -*- coding: utf-8 -*-
"""
Created on Thu Aug 14 16:11:39 2025

@author: JGL
"""

import matplotlib.pyplot as plt
import numpy as np

#aca hago como la primera que lo hicimos en clase
#n = np.arange(0,2*np.pi, 0.1)
#x = np.sin(n)

#aca lo dejo como lo hicimos en clase
fs = 1000 ## si mi fs != a N estoy cambiando mi relacion de 1/deltaf = N.Ts = 1 y cambio el tiempo.
N = 1000   ##esto quiere decir que yo voy a tomar 1000 muestras por segundo. 
deltaF = fs/N
Ts = 1/(N*deltaF)

tiempo = np.arange(0, N * Ts, Ts)
fx =1
A = 1
x = A * np.sin(2 * np.pi * fx * tiempo)
plt.plot(tiempo,x)
plt.show

## aca lo hago como lo piden en el moodle
def mi_funcion_sen(vmax=1, dc=0, ff=1, ph=0, nn=1000, fs=1000):
    Ts = 1/fs
    tt = np.arange(nn) * Ts           # vector de tiempo
    xx = dc + vmax * np.sin(2*np.pi*ff*tt + ph)  # se침al senoidal
    return tt, xx
    
tt, xx = mi_funcion_sen(vmax=1, dc=0, ff=1001, ph=0, nn=N, fs=fs)
plt.plot(tt, xx)
plt.show()
```

# Bonus
Aca realice el bonus que pedia graficar distintos tipos de se침ales para ver como se comporta la se침al.Cuando la frecuencia de la se침al alcanza 洧녭洧/2, aparece el l칤mite de Nyquist: la se침al ya no puede ser reconstruida correctamente. Para frecuencias mayores, debido a la periodicidad de la transformada discreta, la se침al se pliega (aliasing) y una frecuencia como 1001 Hz se observa digitalmente como si fuera 1 Hz.

```python
import numpy as np
import matplotlib.pyplot as plt

fs = 1000
N = 1000
Ts = 1/fs
tiempo = np.arange(0, N*Ts, Ts)
A = 1

x1 = A * np.sin(2*np.pi*500*tiempo)
x2 = A * np.sin(2*np.pi*999*tiempo)
x3 = A * np.sin(2*np.pi*1001*tiempo)
x4 = A * np.sin(2*np.pi*2001*tiempo)

plt.subplot(2,2,1)
plt.plot(tiempo,x1)
plt.title("500 Hz")

plt.subplot(2,2,2)
plt.plot(tiempo,x2)
plt.title("999 Hz")

plt.subplot(2,2,3)
plt.plot(tiempo,x3)
plt.title("1001 Hz")

plt.subplot(2,2,4)
plt.plot(tiempo,x4)
plt.title("2001 Hz")

plt.tight_layout()  # ajusta los t칤tulos y ejes
plt.show()
```
