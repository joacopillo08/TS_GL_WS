## 1. Cálculo de la función de transferencia 

### Cálculo de $T(z)_a$

Partimos de la ecuación en diferencias

$$y(n)=x(n−3)+x(n−2)+x(n−1)+x(n)$$

Aplicando la transformada $z$ (condiciones iniciales nulas) y usando que $x(n-k)\longleftrightarrow z^{-k}X(z):$

$$Y(z)=X(z)z^{-3}+X(z)z^{-2}+X(z)z^{-1}+X(z)$$
$$Y(z)=X(z)(z^{-3}+z^{-2}+z^{-1}+1)$$
Por lo tanto la función de transferencia es
$$T(z)=\frac{Y(z)}{X(z)}=z^{-3}+z^{-2}+z^{-1}+1$$
Factorizada como
$$T(z)=z^{-3}(z^0+z^1+z^2+z^3)=\frac{1+z^1+z^2+z^3}{z^3}$$

Ahora evaluamos en la frecuencia $z=e^{j\omega}$:
$$T(e^{j\omega})=\frac{e^{j0}+e^{j\omega}+e^{2j\omega}+e^{3j\omega}}{e^{3j\omega}}$$

$$T(e^{j\omega})=\frac{e^{j\frac{3\omega}{2}}(e^{-j\frac{3\omega}{2}}+e^{-j\frac{1\omega}{2}}+e^{j\frac{1\omega}{2}}+e^{j\frac{3\omega}{2}})}{e^{3j\omega}}=e^{-j\frac{3\omega}{2}}$$

Usando la fórmula de la suma geométrica:

$$T(e^{j\omega})=e^{-j\frac{3\omega}{2}}\frac{sen(2\omega)}{sen(\frac{\omega}{2})}$$

De forma equivalente para ver la parte real:
$$T(e^{j\omega})=e^{-j\frac{3\omega}{2}}(2cos(3\omega/2)+2cos(\omega/2))$$

**Módulo y fase**
$$|T(e^{j\omega})|=|2cos(3\omega/2)+2cos(\omega/2)|=|\frac{sen(2\omega)}{sen(\frac{\omega}{2})}|$$
$$\angle T(e^{j\omega})=-\frac{3}{2}\omega$$




### Cálculo de $T(z)_b$

Partiendo de la ecuación en diferencias:
$$y(n)=x(n-4)+x(n-3)+x(n-2)+x(n-1)+x(n)$$
Aplicando la transformada $z$ (condiciones iniciales nulas) y usando que $x(n-k)\longleftrightarrow z^{-k}X(z):$
$$Y(z)=X(z)(z^{-4}+z^{-3}+z^{-2}+z^{-1}+z^0)$$
Por lo tanto la función de transferencia es:
$$T(z)=\frac{Y(z)}{X(z)}=\frac{z^4+z^3+z^2+z^1+z^0}{z^4}$$
Ahora evaluamos en $z=e^{j\omega}$ para obtener la respuesta en frecuencia:
$$T(ej^{j\omega})=\frac{e^{4j\omega}+e^{3j\omega}+e^{2j\omega}+e^{j\omega}+e^0}{e^{4j\omega}}=\frac{e^{2j\omega}(e^{2j\omega}+e^{j\omega}+e^0+e^{-j\omega}+e^{-2j\omega})}{e^{4j\omega}}$$
Usando las identidades de cosenos esto se puede escribir también como:
$$T(e^{j\omega})=e^{-2j\omega}(2cos(2\omega)+2cos(\omega)+1)$$

**Módulo y fase**

$$|T(e^{j\omega})|= 2|cos(2\omega)+cos(\omega)|+1$$
$$\angle T(e^{j\omega})=-2\omega$$



### Cálculo de $T(z)_c$

# 3) Simulación de la respuesta en frecuencia


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal

fs = 1000


def resp_en_w(b, a, nombre = 'Filtro', fs = 1000):
    w, h= signal.freqz(b = b, a = a, worN = 1024, fs = fs) #calcula rta en frq del filtro, devuelve w y vector de salida (h es numero complejo)

    fase = np.unwrap(np.angle(h)) #unwrap hace grafico continuo

    #w_rad = w * (fs / (2 * np.pi))
    gd = -np.diff(fase) / np.diff(2*np.pi*w) * fs   # τg [muestras]

    z, p, k =  signal.tf2zpk(b, a) #ubicacion de polos y ceros, z=ubicacion de ceros(=0), p=ubicacion polos, k
    
    # Magnitud
    plt.figure(figsize=(10,8))
    plt.subplot(2,2,1)
    plt.plot(w, 20*np.log10(abs(h)+1e-12))
    plt.title(f'{nombre}: Magnitud')
    plt.xlabel('Frecuencia [Hz]')
    plt.ylabel('|H(e^{jω})| [dB]')
    plt.grid(True, which='both', ls=':')
    
    # Fase
    plt.subplot(2,2,2)
    plt.plot(w, fase)
    plt.title(f'{nombre}: Fase')
    plt.xlabel('Frecuencia [Hz]')
    plt.ylabel('Fase [rad]')
    plt.grid(True, which='both', ls=':')
    
    # Retardo de grupo
    plt.subplot(2,2,3)
    plt.plot(w[:-1], gd)
    plt.title(f'{nombre}: Retardo de grupo')
    plt.xlabel('Frecuencia [Hz]')
    plt.ylabel('τg [muestras]')
    plt.grid(True, which='both', ls=':')
    
    plt.suptitle(f'Respuesta en frecuencia – {nombre}')
    plt.tight_layout()
    plt.show()

    

b1 = ([1,1,1,1])
a1 = ([1])

b2 = ([1,1,1,1,1])
a2 = ([1])

b3 = ([1,-1])
a3 = ([1])

b4 = ([1,0,-1])
a4 = ([1])

funcionA = resp_en_w(b = b1, a = a1, nombre = 'Filtro a')
funcionB = resp_en_w(b = b2, a = a2, nombre = 'Filtro b')
funcionC = resp_en_w(b = b3, a = a3, nombre = 'Filtro c')
funcionD = resp_en_w(b = b4, a = a4, nombre = 'Filtro d')
```

## Análisis de los gráficos:

### Filtro a
La respuesta en magnitud muestra un comportamiento típico de un filtro pasabajos, lo que genera múltiples ceros en el eje unitario y atenuaciones fuertes a frecuencias armónicas del muestreo. La fase es lineal a tramos, lo que indica un retardo constante entre las muestras. El retardo de grupo presenta picos negativos por el modo en que se calcula numéricamente, pero conceptualmente corresponde a un retardo medio de 2 muestras (la mitad de la ventana).

### Filtro b
Al tener cinco coeficientes iguales, la magnitud muestra una respuesta más selectiva con lóbulos más estrechos y mayor atenuación en las frecuencias donde aparecen ceros. Esto implica un mayor suavizado de la señal. La fase mantiene la linealidad esperada para un filtro FIR simétrico, y el retardo de grupo es constante y de unas 2 muestras y media, coherente con el orden del filtro.

La fase es lineal, con pendiente negativa, lo que implica un retardo constante de media muestra. El retardo de grupo se mantiene constante, reflejando una respuesta de fase lineal y por tanto sin distorsión temporal.
